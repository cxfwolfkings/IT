我打算通过基金定投（每月3000，年盈利18%，20年），达成财务自由的目标。

今天是2021年2月24，第一年，我的收益目标是3720元。

本来一切都很美好，短短一个多月就赚了2000元，不过突然之间形势急转直下，基金跌跌不休，不但盈利全失，还亏损了本金。

究其原因，我在这段时间的投资中主要犯了以下几个毛病：

1. 没有止损！任何投资都是需要止损的，为什么呢？保住本金，等待更好的入场机会。基金虽然从长远来看几乎是稳定盈利的，但是盈利多少才是问题关键，一旦情况有变，早点减仓或清仓，不但能避免更多损失，还能在更好的时机，用较多的筹码杀回市场，收复失地，进而大赚一笔。

   同时，还有几点要注意：
   
   - 不要让盈利变成损失
   - 当出现不好的信号时，不要猜测市场走向，一定要稳中求胜（本次已经出现抱团基金瓦解的信号，我没有及时清仓，导致损失）
   

2. 不要靠猜测作决定，一定要花时间研究，按照投资策略来操作。我最近的操作，毫无逻辑可言，只是猜测哪些基金能赚钱，完全不经过研究就买入。更坏的情况是，当基金亏损时，我也想当然地通过再投入资金来降低成本，结果越亏越多。

   选择一定是基于某种策略，不能一拍脑瓜就决定，下面几点以后一定注意了：
   
   - 一定要大胆买入低估值的基金，谨慎买入中估值的基金，不买高估值的基金。指数基金可以使用博格公式法进行估值，主动型基金可以根据10大重仓股的市盈率（市净率）进行估值
   - 不要轻易补仓。补仓之前一定要研究清楚当前市场的走向，看不清，则不动，一定要耐住性子！


下面来说说，为什么基金经理不换股，其中的逻辑是什么？（只是个人的一点感想）

我当前投资的这只基金2020年年化收益率达到惊人的130%，查看其重仓股，基本都是各个行业的龙头。为什么在最近龙头大跌的情况下，还不换股呢？

其实从其过往的收益率看，该基金一定是在一个比较低的价位买入这些龙头股的，这就说明，它的护城河比较宽。虽然最近龙头股大跌，但是龙头毕竟是龙头，行业的领先地位没变，以后一定是会逐渐收复失地的。而又因为，这种往年收益率很高的基金已经有了很高的安全边际，即使一段时间行情大跌，它们仍然能保持不错的收益率，就像我入手的这只基金，经过这几天连续的下跌，年度收益率还是在100%以上，那么基金经理还有什么好慌的呢？反而，在这种时候，如果将手里的龙头股换掉，一来行情震荡，未必能选中其它赚钱的股票；二来，等龙头发威的时候，不一定能抓住好的买入时机。一动不如一静，不如稳坐钓鱼台，静观其变。

回到我们自己身上，因为我已经是在估值很高的时候，才买入该基金的了，所以没什么安全边际，一有这种连续下跌的行情（市场上还有抱团基金瓦解的消息），我就扛不住了。

所以啊，综上所述，高估值的基金尽量不要买，如果在确定没有风险的时候为了快速赚钱买入了，就一定要做好止损。目标收益不可以定太高，市场上一旦有不利的消息，同时行情下跌，就立刻清仓一半，若第二天继续下跌，再清仓一半，连跌5天，基本上就是真的转势的信号了，赶紧将资金全部撤出来吧，别忘了，留10份在基金里，等待好的入场时机，再次杀回来。
