指数基金 VS 混合基金

指数基金使用被动策略跟随指数，因此市场未来趋势就是选择的前提。一般可以通过历史分位法来评估，市盈率和市净率的历史分位都小于20%时，是很好的入场时机，可以投入初始资金，然后开始定投；历史分位超过50%时，可以继续持有，但是不再定投，多余的资金可以投资其它标的，或者放在手中，当满足定投条件时，增加定投额，平均到后面3次的定投金额中。当历史分位超过80%时，则可以逐步清仓，也是按月清仓。清道30%的仓位时，按兵不动，吃一波“疯狂”的收益，一旦从高点跌幅超过6%，并且市场中有不利消息，则全部清仓；如果没有利空消息，则保留10%的仓位，后续操作雷同。

混合基金由基金经理主动选股，所以基金经理的能力和基金的选股策略就是关键所在。在我国，指数基金实现一年40%，不太现实，但是主动型基金，只要市场环境好，大概率是能够实现的。估算混合基金的未来收益，可以通过其持有的重仓股。一般此类基金重仓的都是行业龙头，而行业龙头的表现可以说是大盘中最亮眼的（至少也是处于上游）。所以，定投混合基金，是很好的赚钱策略。不过因为龙头公司发展较快，难免会有估值过高，回调的时候。此时就要像操作股票一样，及时调整仓位，宁可少赚，不要亏损，等到回调到位，再逐步加仓。如此操作，不但能获得较高收益，持仓成本也能逐步降低。
