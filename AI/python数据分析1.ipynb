```python
import matplotlib
'导入pyplot'
from matplotlib import pyplot as plt
from matplotlib import font_manager

'''
font = {'family':'MicroSoft YaHei','weight':'bold','size':'larger'}
matplotlib.rc("font",**font)
matplotlib.rc("font",family="MicroSoft YaHei",weight="bold")
'''
my_font=font_manager.FontProperties(fname="C:/Windows/Fonts/msyhbd.ttc")

'''
设置图片大小
figure：图形图标的意思，在这里指我们画的图
'''
fig = plt.figure(figsize=(20,8),dpi=80)

'数据在x轴的位置'
x = range(2,26,2)
'数据在y轴的位置'
y = [15,13,14.5,17,20,25,26,26,24,22,18,15]
y2 = [12,15,12.5,18,24,16,16,18,22,26,15,25]

'设置坐标轴刻度'
_xtick_labels = ["{}层".format(i) for i in range(2,14)]
_xtick_labels += ["{}级".format(i) for i in range(14,25)]
plt.xticks(range(2,27),_xtick_labels,rotation = 60,fontproperties=my_font)
plt.yticks([i/2 for i in range(min(y)*2,max(y)*2+1)])

'通过plot绘制折线图'
plt.plot(x,y,label="自己",color="red",linewidth=4)
plt.plot(x,y2,label="宠物",linestyle="--")

'添加描述信息'
plt.xlabel("等级（层次）")
plt.ylabel("力量")
plt.title("测试力量曲线图")

'绘制网格'
plt.grid(alpha=0.4)

'添加图例'
plt.legend(prop=my_font)

'保存矢量图，不会失真'
'plt.savefig("t1.svg")'

'展示图形'
plt.show()
```
